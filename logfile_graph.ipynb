```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
with open("logfile.txt", "r") as ins:
    arr = []
    for line in ins:
        arr.append(line)
if not os.path.exists('train_data_graphs'):
        os.makedirs('train_data_graphs')
```

```python
dt = np.array(arr)
k = dt
```

```python
for i in range(0,len(dt) -1 ):
    print(dt[i])
```

```python
E_P = []
V_A = []
T_A = []
V_L = []
T_L = []
for i in range(0,len(dt) - 1):
    
    
    #for Epoch:
    
    temp = dt[i].split(',')
    temp1 = temp[0].split(' ')
    #print( int(temp1[1]) )
    E_P.append( int(temp1[1]) )
    
    
    #for V_A:
    
    temp2 = temp[1][0:-1]
    temp3 = temp2.split(' ')
    temp3 = temp3[-1]
    #print ( float(temp3) )
    V_A.append( float(temp3) )
    
    #for T_A:
    
    temp4 = temp[0][0:-1]
    temp5 = temp4.split(' ')
    temp5 = temp5[-1]
    #print ( float(temp5) )
    T_A.append( float(temp5) )
    
    #for Validation Loss:
    
    temp6 = temp[2].split(' ')
    temp7 = temp6[-1]
    #print( float(temp7) )
    V_L.append( float(temp7))
    
    #for train_loss :
    
    temp8 = temp[3][0:-1] 
    temp9 = temp8.split(' ')
    temp10 = temp9[-1]
    #print( float(temp10) )
    T_L.append( float(temp10) )
    
    
    print("{}    {}    {}    {}    {}".format( E_P[i], V_A[i], T_A[i], V_L[i], T_L[i]))
    
    
    
    
   
        
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,T_A,lw = 1)
plt.xlabel('epoch')
plt.ylabel('training accuracy')
plt.title('training accuracy vs epochs')
plt.savefig('train_data_graphs/T_A_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,V_A,lw= 1)
plt.xlabel('epoch')
plt.ylabel('validation accuracy')
plt.title('validation accuracy vs epochs')
plt.savefig('train_data_graphs/V_A_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,V_L,lw= 1)
plt.xlabel('epoch')
plt.ylabel('validation loss')
plt.title('validation loss vs epochs')
plt.savefig('train_data_graphs/V_L_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,T_L,lw= 1)
plt.xlabel('epoch')
plt.ylabel('training loss')
plt.title('training loss vs epochs')
plt.savefig('train_data_graphs/T_L_vs_epochs.png')
```

```python
z = np.polyfit(E_P, V_A, 3)


f = np.poly1d( z )
print( f )
lr  = np.zeros( 1080 )
lr1 = np.zeros( 1080 )
for i in range(0, 1080):
    lr[i] = f(i)
    lr1[i] = i
```

```python
plt.figure(figsize=(30,10))

plt.plot(lr1, lr, lw = 3)
plt.xlabel('epoch')
plt.ylabel('validation accuracy')
plt.title('validation accuracy vs epochs best fir curve')
plt.savefig('train_data_graphs/V_A_vs_epochs_best_fir.png')
```
